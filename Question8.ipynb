# Question 8
- Find out how many unusual words people speak in chat

```python
from nltk.corpus import nps_chat
from nltk.corpus import words

all_post = nps_chat.fileids() #Getting all files in a list
word_list = words.words()  # Getting all words in a list
chat_words = []    
all_words = []
unusual_words =[]

for x in word_list:            #Iterating over the list of words and add it in another list with all lowercase
   all_words.append(x.lower())
 
for x in all_post:             #Checking if all words of all files of nps_chat is alphabet and appending it in a list
  for y in nps_chat.words(x): 
    if y.isalpha(): 
        chat_words.append(y.lower())
    
for i in chat_words:            #Checking what  all words of nps_chat is nlot there in common words list
    if i not in all_words:
        unusual_words.append(i)
        
print("unusual words: ",unusual_words) #Printing the unusual words
        
```

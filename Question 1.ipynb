# Question 1
- Find out how many words and how many unique words in file 'austen-persuasion.txt'.
- Calculate the frequency for each words and show the top 10 most frequent words and their frequency.

```python
from nltk.corpus import gutenberg
files = gutenberg.fileids() # Get all files and put them in list.
total_words =[] #Initialize an array
print("files in gutenberg:", files)

all_words = gutenberg.words('austen-persuasion.txt') # Get all words from file
for x in all_words:
    total_words.append(x) # add words to list
print("Total words in file :", len(total_words))
words_freq = {}      # Initilize dictionary for word frequency
top_10 ={}           # Initilize dictionary for top 10 high frequency words
sum=0
```

```python
for i in total_words:         # check word frequency and add in a dictionary
        if i in words_freq:
            words_freq[i]+=1
        else:
            words_freq[i]=1
            
```

```python
print("Words with their frequencies in file :", words_freq) # Print word frequency

```

```python
top_10 = sorted(words_freq, key= words_freq.get, reverse=True)[:10] #Sort word frequency to get top 10 words
print('Top 10 words with their frequency')
for i in top_10:
    print('key:', i ,'value:' ,words_freq[i])     #Print top 10 words
```

```python
for word, freq in words_freq.items():  #Find all uniques words
    if freq == 1:
        sum+=1

print("Total unique words :", sum)        

```

- Create a folder structure to store them into these separate folders and read all these json files.
- Create a hierarchy of folder structure. 

```python
import json
import os
import shutil
import glob
import xlwt
```

```python
if not os.path.exists('Data Processed'): #making the folder Data Processed in location if folder doesn't exist
    os.mkdir('Data Processed')
    
path ='Data Processed'        #root path of folder structures
rootdir='Data'                #path where all json are kept
```

```python
for infile in glob.glob(os.path.join(rootdir, '*.json')):
       with open(infile) as filedata: 
            data = json.load(filedata) 
            term = data["term"]       #taking out the term name from json
            country_name = data["location"]["country"]        #taking out the country name from json
            filepath = os.path.join(path, country_name) #making the folder under Data Processed folder with folder name countryName found from json
            if not os.path.exists(filepath):         
                os.mkdir(filepath)
    
            termpath = os.path.join(filepath, term) #making the folder under data folder with folder name as term found from json
            if not os.path.exists(termpath):         
                os.mkdir(termpath)    
        
            shutil.copy(infile, termpath)    #moving the json file inside filepath  
            print(termpath)                  #printing the file path of new moved location
            filedata.close()
```

```python
book = xlwt.Workbook(encoding="utf-8")      #Creating an excel sheet to load data  
sheet = book.add_sheet("resturant")
sheet.write(0, 0,"Name of restaurant")       #Creating title of the excel sheet
sheet.write(0, 1,"City")
sheet.write(0, 2,"Country")
sheet.write(0, 3,"Day of week")
sheet.write(0, 4,"Start Time Hour")
sheet.write(0, 5,"Start Time Minute")
sheet.write(0, 6,"End time Hour")
sheet.write(0, 7,"End Time Minute")
row = 1                       #initializing row and col to iterate over the excel sheet from second row       
col = 0 

```

```python
for infile in glob.glob(os.path.join(rootdir, '*.json')):
       with open(infile) as filedata: 
            d1 = json.load(filedata)                                 #extracting data from json
            if data["term"] == "restaurants":
                if "hours" in d1: 
                    for d2 in d1["hours"]:       # iterating over the json
                        for d3 in d2["open"]:
                            sheet.write(row,col,d1["name"])                  #writing the data in an excel sheet 
                            sheet.write(row,col+1,d1["location"]["city"])
                            sheet.write(row,col+2,d1["location"]["country"])
                            sheet.write(row,col+3,d3["day"]) 
                            sheet.write(row,col+4,d3["start"][:2]) 
                            sheet.write(row,col+5,d3["start"][-2:]) 
                            sheet.write(row,col+6,d3["end"][:2]) 
                            sheet.write(row,col+7,d3["end"][-2:]) 
                            row+=1

book.save("python_spreadsheet.xls")             #Saving the excel sheet 
filedata.close()  
```
